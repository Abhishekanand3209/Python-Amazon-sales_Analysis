# Group BY

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=S_Qty)
```

```python
sns.countplot(data=df, x='Courier Status',hue='Status')
#courier Status
```

Note:the majority of order are shipped through the courier.


```python
#histograph
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins =20, edgecolor='Green')
plt.xticks(rotation=90)
plt.show()
```

Note : From above graph, I can easily most of the buyers are T-Shirt

```python
# using pie chart, we can check B2B data
B2B_Check = df['B2B'].value_counts()

# pie- chart
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')

#plot.axis('equal')
plt.show()
```

Note : From pie chart,Ican see 99.3% are retailer and 0.7% are B2B buyers

```python
 #showing scatter plot
x_data= df['Category']
y_data= df['Size']

#PLoting for scatter
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show
```

```python
# plot cities by state distribution
plt.figure(figsize=(15,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show
```

```python
#finding top state
top_10_state=df['ship-state'].value_counts().head(15)
plt.figure(figsize=(15,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show
```

Note : from above graph we can easily see, most of the buyera are from Maharashtra State
